```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = np.arange(-5,5,.1)
```

```python
def func(x):
    x = 1/(1 + np.exp(-x))
    x = np.square(x)
    
    return x
```

```python
def func1(x):
    x = 1/(1 + np.exp(-x))
    x = np.square(x)
    x = -np.log(x)
    
    return x
```

```python
def func2(x):
    x = 1/(1 + np.exp(-x))
    x = np.square(x)
    x = -np.log(1-x)
    
    return x
```

```python
y = func(x)
```

```python
y1 = func1(x)
```

```python
y2 = func2(x)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y1)
plt.show()
```

```python
plt.plot(x,y2)
plt.show()
```
