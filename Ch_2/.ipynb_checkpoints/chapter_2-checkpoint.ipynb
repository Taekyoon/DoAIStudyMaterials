## Get the Data

데이터는 책에 있는 내용과 다른 URL에서 다운이 가능하다.

```python
DATASETS_URL = "https://github.com/ageron/handson-ml/raw/master/datasets"
```

아래 코드는 데이터를 받을 위치를 지정하여 다운을 받는다.

```python
import os
import tarfile
from six.moves import urllib

HOUSING_PATH = "datasets/housing"
HOUSING_URL = DATASETS_URL + "/housing/housing.tgz"

def fetch_housing_data(housing_url=HOUSING_URL, housing_path=HOUSING_PATH):
    if not os.path.exists(housing_path):
        os.makedirs(housing_path)
    tgz_path = os.path.join(housing_path, "housing.tgz")
    urllib.request.urlretrieve(housing_url, tgz_path)
    housing_tgz = tarfile.open(tgz_path)
    housing_tgz.extractall(path=housing_path)
    housing_tgz.close()
```

```python
fetch_housing_data()
```

### Split Whole Data into Train and Test Data

일반적으로 데이터는 학습 60% 교차검증 20% 테스트 20%로 구분하여 모델을 만든다.

```python
import numpy as np
import numpy.random as rnd
rnd.seed(42) # to make this notebook's output identical at every run

# 학습과 테스트 데이터를 분리하는 함수
def split_train_test(data, test_ratio):
    shuffled_indices = rnd.permutation(len(data)) # 고정된 데이터 형태로 학습과 테스트를 하지않기 위해 가진 데이터를 섞는다.
    test_set_size = int(len(data) * test_ratio) # 테스트 데이터 비율은 보통 20%이다.
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    return data.iloc[train_indices], data.iloc[test_indices]
```

```python
train_set, test_set = split_train_test(housing, 0.2)
print('train: ', len(train_set), 'test: ', len(test_set))
```

```python
import hashlib

# 추가적인 데이터가 있는 경우엔 학습과 테스트를 할 데이터에 대해 섞을 수 있도록 하여야 한다.
# 기존에 가지고 있는 데이터 끝 부분에 있는 값을 가지고 해쉬값을 만들어 인덱스를 생성하고 학습데이터와 테스트데이터를 구분한다.
def test_set_check(identifier, test_ratio, hash):
    return bytearray(hash(np.int64(identifier)).digest())[-1] < 256 * test_ratio

def split_train_test_by_id(data, test_ratio, id_column, hash=hashlib.md5):
    ids = data[id_column]
    in_test_set = ids.apply(lambda id_: test_set_check(id_, test_ratio, hash))
    return data.loc[~in_test_set], data.loc[in_test_set]
```

```python
housing_with_id = housing.reset_index()   # adds an `index` column
train_set, test_set = split_train_test_by_id(housing_with_id, 0.2, "index")
test_set.head()
```

index 외에 다른 데이터를 활용하여 인덱스를 만들어 데이터를 구분할 수 있다.

```python
# index를 생성하기 위한 요소로 위도와 경도를 사용한다.
housing_with_id['id'] = housing['longitude'] * 1000 + housing['latitude']
train_set, test_set = split_train_test_by_id(housing_with_id, 0.2, "id")
test_set.head()
```

### Prepare for Sampling Data

데이터가 충분히 많은 경우에는 Randon Samping이 가능하다. 하지만 데이터가 적은 경우에는 Sampling Bias가 생길 가능성이 있다.
특히 데이터의 층(수의 분포)이 많아질 수록 데이터 모델링은 Biased 될 가능성이 있다. 현 데이터에서는 ceiling을 두고 5개의 계급으로 구분해 두도록 한다.
이 데이터에서는 수입의 중간값이 집 가격에 대한 중간값을 예측하는데 중요한 역할을 하므로 계급에 대한 구분 기준을 수입을 토대로 한다. 

```python
housing["income_cat"] = np.ceil(housing["median_income"] / 1.5)
housing["income_cat"].where(housing["income_cat"] < 5, 5.0, inplace=True)
housing["income_cat"].value_counts() / len(housing)
```

```python
housing["income_cat"].hist()
```

위의 그래프에서 보이듯이 데이터의 분포는 왼쪽으로 편향되어 분포가 되어있다.

```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(housing, housing["income_cat"]):
    strat_train_set = housing.loc[train_index]
    strat_test_set = housing.loc[test_index]
```

```python
def income_cat_proportions(data):
    return data["income_cat"].value_counts() / len(data)

train_set, test_set = train_test_split(housing, test_size=0.2, random_state=42)

compare_props = pd.DataFrame({
    "Overall": income_cat_proportions(housing),
    "Stratified": income_cat_proportions(strat_test_set),
    "Random": income_cat_proportions(test_set),
}).sort_index()
compare_props["Rand. %error"] = 100 * compare_props["Random"] / compare_props["Overall"] - 100
compare_props["Strat. %error"] = 100 * compare_props["Stratified"] / compare_props["Overall"] - 100
```

아래 표를 통해서 Random Sampling과 Stratified Samling의 오류가 얼마나 차이가 나는지 확인 할 수 있다. 
수치에서 볼 수 있듯이 Stratified Samping이 전체 데이터와 비교하였을 때 계층에 따라 고르게 분포되어 Sampling이 된 것을 확인 할 수 있다.

```python
compare_props
```

원래 데이터 형태로 돌리기 위해 계급에 관한 attribute를 제거한다.

```python
for set in (strat_train_set, strat_test_set):
    set.drop("income_cat", axis=1, inplace=True)
```

Data Cleaning을 위해 처리하는 방법은 3가지로 구분이 된다.

- 해당 row 데이터를 제거한다.
- 전체 attribute 데이터를 제거한다.

랜덤변수의 분포는 해당된 데이터의 평균과 분산값을 활용하여 생성을 하도록 하였다.
이 경우에 대해서는 실행 전과 후의 데이터 분포를 확인 할 필요가 있다.

```python
housing_copy_target = housing.copy()

# get average, std, and number of NaN values in housing_copy_compare
average_total_bedrooms = housing_copy_compare['total_bedrooms'].mean()
std_total_bedrooms       = housing_copy_compare['total_bedrooms'].std()
count_nan_total_bedrooms = housing_copy_compare['total_bedrooms'].isnull().sum()
# generate random numbers between (mean - std) & (mean + std)
rand_1 = np.random.randint(average_total_bedrooms - std_total_bedrooms, average_total_bedrooms + std_total_bedrooms, size = count_nan_total_bedrooms)

housing_copy_target['total_bedrooms'][np.isnan(housing_copy_target['total_bedrooms'])] = rand_1
```

기존의 total_bedrooms의 데이터 분포 그래프와 비교해보았을 때 크게 별 다른 차이를 보이지는 않는다.

### Handling Text and Categorical Attributes

### Custom Transformers

지금까지는 데이터를 머신러닝에 학습시킬 데이터를 정제하는 과정을 거쳤다. 
이 과정을 보다 편리하게 관리하기 위해 duck typing을 한다. 
아래 코드 내용과 같이 TransformerMixin을 상속한 클래스를 생성할 수 있다. 여기에서 반드시 구현해야 하는 함수는 fit, transform, fit_transform 함수가 있다. 
(이 책 저자의 의도는 이를 활용하여 보다 편리한 파이프라이닝을 하기 위함인 것 같다.)

```python
from sklearn.base import BaseEstimator, TransformerMixin

rooms_ix, bedrooms_ix, population_ix, household_ix = 3, 4, 5, 6

# 각 attribute 값에 대한 조합을 처리하는 정제클래스
class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def __init__(self, add_bedrooms_per_room = True): # no *args or **kargs
        self.add_bedrooms_per_room = add_bedrooms_per_room
    def fit(self, X, y=None):
        return self  # nothing else to do
    def transform(self, X, y=None):
        rooms_per_household = X[:, rooms_ix] / X[:, household_ix]
        population_per_household = X[:, population_ix] / X[:, household_ix]
        if self.add_bedrooms_per_room:
            bedrooms_per_room = X[:, bedrooms_ix] / X[:, rooms_ix]
            return np.c_[X, rooms_per_household, population_per_household, bedrooms_per_room]
        else:
            return np.c_[X, rooms_per_household, population_per_household]

attr_adder = CombinedAttributesAdder(add_bedrooms_per_room=False)
housing_extra_attribs = attr_adder.transform(housing.values)

housing_extra_attribs = pd.DataFrame(housing_extra_attribs, columns=list(housing.columns)+["rooms_per_household", "population_per_household"])
housing_extra_attribs.head()
```

파이프라이닝은 데이터 정제과정을 하는데 있어 편리하게 해주는 역할을 한다. 각 정제과정을 연결해두어 원하고자 하는 데이터 결과물을 얻도록 하는 것이 목적이다. 파이프라이닝의 다른 장점은 각 정제과정의 모듈을 여러가지로 조합하여 결과물을 얻도록 할 수 있다는 점에 있다.

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

num_pipeline = Pipeline([
        ('imputer', Imputer(strategy="median")),
        ('attribs_adder', CombinedAttributesAdder()),
        ('std_scaler', StandardScaler()),
    ])

num_pipeline.fit_transform(housing_num)
```

최종적으로 머신러닝 학습을 위한 데이터 준비는 아래와 같다.

```python
from sklearn.pipeline import FeatureUnion

# 해당 attribute에 대한 값들을 추출하는 정제클래스
class DataFrameSelector(BaseEstimator, TransformerMixin):
    def __init__(self, attribute_names):
        self.attribute_names = attribute_names
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        return X[self.attribute_names].values

num_attribs = list(housing_num)
cat_attribs = ["ocean_proximity"]

# 숫자 데이터에 대한 처리를 하는 파이프라인
num_pipeline = Pipeline([
        ('selector', DataFrameSelector(num_attribs)), # 각 attribute에 대한 데이터를 추출한다.
        ('imputer', Imputer(strategy="median")),      # NaN인 데이터에 대한 fill을 해준다.
        ('attribs_adder', CombinedAttributesAdder()), # attribute와의 결합한 feature들을 추가한다.
        ('std_scaler', StandardScaler()),             # Feature Scaling을 한다.
    ])

# 텍스트 데이터에 대한 처리를 하는 파이프라인
cat_pipeline = Pipeline([
        ('selector', DataFrameSelector(cat_attribs)), # 해당 attrivute에 대한 데이터 추출한다.
        ('label_binarizer', LabelBinarizer()),        # Text 데이터에 대해 One hot vector로 변환한다.
    ])

# 메인 파이프라인인
preparation_pipeline = FeatureUnion(transformer_list=[
        ("num_pipeline", num_pipeline),
        ("cat_pipeline", cat_pipeline),
    ])
```

처리를 한 결과는 아래와 같이 볼 수 있다.

```python
housing_prepared = preparation_pipeline.fit_transform(housing)
housing_prepared
```

### Cross Validation

실제 머신러닝 학습과정에서는 학습데이터와 교차검증데이터를 활용하여 모델을 튜닝한다.
아래 cross_val_score 함수는 머신러닝 학습에 대한 교차검증을 해주는 함수이다. K-fold cross-validation을 이용하여 검증을 할 수 있다. K-fold 교차검증은 전체 데이터를 K개의 fold로 나누어 1개 fold를 검증 데이터로 지정하고 나머지 9개 fold를 학습 데이터로 지정하는 방식으로 검증을 하도록 한다. 함수에서 변수 K를 지정하는 방법은 파라미터 cv에 원하는 K 수를 입력하면 된다. 아래 코드에서는 10으로 입력하여 K값은 10이다.

```python
from sklearn.model_selection import cross_val_score

tree_scores = cross_val_score(tree_reg, housing_prepared, housing_labels,
                              scoring="neg_mean_squared_error", cv=10)
tree_rmse_scores = np.sqrt(-tree_scores)
```

```python
def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())

display_scores(tree_rmse_scores)
```

```python
lin_scores = cross_val_score(lin_reg, housing_prepared, housing_labels,
                             scoring="neg_mean_squared_error", cv=10)
lin_rmse_scores = np.sqrt(-lin_scores)
display_scores(lin_rmse_scores)
```

### Fine Tune Your Model 

머신러닝은 hyperparameter를 조작하여 성능을 조절하여 성능을 높힐 수 있다. 이를 보다 쉽게 해줄 수 있는 것이 GridSearchCV와 RandomizeCV이다. 
GridSearchCV의 경우는 모델에 해당하는 파라미터를 지정하여 각 조합의 결과를 확인할 수 있도록 하고 RandomizeCV의 경우는 파라미터를 랜덤으로 지정하여 확인할 수 있는 방식이다.
RandomizeCV의 경우는 hyperparameter로 지정하기에 경우가 많은 경우에 사용한다.

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor

# n_estimators는 RandomForest에서 사용할 트리의 개수이다.
# max_features는 한 노드에서 split될 feature의 개수이다.
# bootstrap은 bootstrap을 활용하는지에 대한 여부를 확인하는 것이다.
param_grid = [
        {'n_estimators': [3, 10, 30], 'max_features': [2, 4, 6, 8]},
        {'bootstrap': [False], 'n_estimators': [3, 10], 'max_features': [2, 3, 4]},
    ]

forest_reg = RandomForestRegressor()
grid_search = GridSearchCV(forest_reg, param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(housing_prepared, housing_labels)
```

```python
grid_search.best_params_
```

```python
cvres = grid_search.cv_results_
for mean_score, params in zip(cvres["mean_test_score"], cvres["params"]):
    print(np.sqrt(-mean_score), params)
```

```python
from sklearn.model_selection import RandomizedSearchCV

# Grid
param_distribs = {
        'bootstrap': [True, False],
        'n_estimators': range(1,30),
        'max_features': range(2,13),
    }

rf_reg = RandomForestRegressor()
rnd_search = RandomizedSearchCV(rf_reg, param_distributions=param_distribs,
                                n_iter=50, cv=5, scoring='neg_mean_squared_error', verbose=2, n_jobs=4)
rnd_search.fit(housing_prepared, housing_labels)
```

```python
negative_mse = rnd_search.best_score_
rmse = np.sqrt(-negative_mse)
rmse
```
